```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
new_df = remove_outliers(new_df, "demand")
new_df = remove_outliers(new_df, "RRP")
new_df = remove_outliers(new_df, "RRP_positive")
```

```python
def get_season(month):
    if month in [12, 1, 2]:  
        return "Summer"
    elif month in [3, 4, 5]:  
        return "Autumn"
    elif month in [6, 7, 8]:  
        return "Winter"
    else:  # Spring (9, 10, 11)
        return "Spring"

df["season"] = df["date"].dt.month.apply(get_season)

seasonal_demand = df.groupby("season")["demand"].mean()

season_order = ["Summer", "Autumn", "Winter", "Spring"]

plt.figure(figsize=(8, 5))
sns.barplot(x=seasonal_demand.index, y=seasonal_demand.values, order=season_order, palette="viridis")
plt.title("Average Electricity Demand by Season in Australia")
plt.xlabel("Season")
plt.ylabel("Demand (MWh)")
plt.grid()
plt.show()
```

```python
# from prophet import Prophet
# df = df.rename(columns={"date": "ds", "demand": "y"})
# df["day_num"] = df["ds"].dt.dayofweek
# df = df[["ds", "y","day_num"]]

# # Using Prophet Model
# model = Prophet()
# model.add_country_holidays(country_name='AU')

# # Adding external regressors
# # model.add_regressor('min_temperature')
# # model.add_regressor('max_temperature')
# # model.add_regressor('solar_exposure')
# # model.add_regressor('RRP')
# # model.add_regressor('school_day')
# model.add_regressor('day_num')
# model.fit(df)

# future=model.make_future_dataframe(periods=30)
# # future["min_temperature"] = df["min_temperature"].iloc[-1]
# # future["max_temperature"] = df["max_temperature"].iloc[-1]
# # future["solar_exposure"] = df["solar_exposure"].iloc[-1]
# # future["RRP"] = df["RRP"].iloc[-1]
# # future["school_day"] = df["school_day"].iloc[-1]
# future["day_num"] = future["ds"].dt.dayofweek
# forecast=model.predict(future)

# plt.figure(figsize=(12, 6))
# plt.plot(df["ds"], df["y"], label="Actual Demand", marker='o', linestyle="-")
# plt.plot(forecast["ds"], forecast["yhat"], label="Predicted Demand", marker='s', linestyle="--", alpha=0.8)
# plt.xlabel("Date")
# plt.ylabel("Demand")
# plt.title("Actual vs. Predicted Demand")
# plt.legend()
# plt.xticks(rotation=45)
# plt.grid(True)
# plt.show()


from prophet import Prophet
import matplotlib.pyplot as plt

# Create a copy to avoid changing the original df
df_prophet = df.copy()

# Prepare columns for Prophet
df_prophet = df_prophet.rename(columns={"date": "ds", "demand": "y"})
df_prophet["day_of_week"] = df_prophet["ds"].dt.dayofweek
df_prophet = df_prophet[["ds", "y", "day_of_week"]]

# Initialize and configure Prophet model
model = Prophet()
model.add_country_holidays(country_name='AU')
model.add_seasonality(name='monthly', period=30.5, fourier_order=8)
model.add_regressor('day_of_week')  

# Fit the model
model.fit(df_prophet)

# Prepare future DataFrame
future = model.make_future_dataframe(periods=365)
future["day_of_week"] = future["ds"].dt.dayofweek

# Predict
forecast = model.predict(future)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(df_prophet["ds"], df_prophet["y"], label="Actual Demand", marker='o', linestyle="-")
plt.plot(forecast["ds"], forecast["yhat"], label="Predicted Demand", marker='s', linestyle="--", alpha=0.8)
plt.xlabel("Date")
plt.ylabel("Demand")
plt.title("Actual vs. Predicted Demand")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()




```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
actual_demand = df_prophet[df_prophet['ds'].isin(forecast['ds'])]['y']
predicted_demand = forecast[forecast['ds'].isin(df_prophet['ds'])]['yhat']
mae = mean_absolute_error(actual_demand, predicted_demand)
rmse = np.sqrt(mean_squared_error(actual_demand, predicted_demand))
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

 The MAE is 5641.466 and the RMSE is 7658.010
 
 We can see from the above plotting as well as from MAE and RMSE values, this Prophet model gives a reasonable prediction of demand based on date seasonality and day of week.

## Using LIGHTGBM Model

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_absolute_error
import lightgbm as lgb
```

```python
df['dayofweek'] = df['date'].dt.dayofweek
df['month'] = df['date'].dt.month
df['school_day'] = (df['school_day'] == 'Y').astype(int)
df['holiday'] = (df['holiday'] == 'Y').astype(int)
```

```python
# Lag features
df['RRP_lag_1'] = df['RRP_positive'].shift(1)
df['RRP_lag_24'] = df['RRP_positive'].shift(24)
df['DEMAND_lag_1'] = df['demand'].shift(1)
```

```python
df['RRP_rolling_mean_24'] = df['RRP_positive'].rolling(window=24).mean()
df['DEMAND_rolling_mean_24'] = df['demand'].rolling(window=24).mean()
```

```python
df.dropna(inplace=True)
```

```python
# Features and target
features = ['dayofweek', 'month', 'school_day', 'holiday',
            'RRP_lag_1', 'RRP_lag_24', 'RRP_rolling_mean_24',
            'DEMAND_lag_1', 'DEMAND_rolling_mean_24']
target = 'RRP'
```

```python
X = df[features]
y = df[target]
```

```python
# TimeSeries split
tscv = TimeSeriesSplit(n_splits=5)
for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = lgb.LGBMRegressor()
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    print("MAE:", mean_absolute_error(y_test, y_pred))

```

```python
# TimeSeries split with comprehensive evaluation
tscv = TimeSeriesSplit(n_splits=5)
mae_scores = []
rmse_scores = []
r2_scores = []

print("=== LightGBM Model - Cross-Validation Results ===")
fold = 1

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    model = lgb.LGBMRegressor(random_state=42, verbose=-1)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mae_fold = mean_absolute_error(y_test, y_pred)
    rmse_fold = np.sqrt(mean_squared_error(y_test, y_pred))
    r2_fold = r2_score(y_test, y_pred)
    
    mae_scores.append(mae_fold)
    rmse_scores.append(rmse_fold)
    r2_scores.append(r2_fold)
    
    print(f"Fold {fold} - MAE: {mae_fold:.2f}, RMSE: {rmse_fold:.2f}, RÂ²: {r2_fold:.4f}")
    fold += 1

print(f"\n=== Overall Performance ===")
print(f"Average MAE: {np.mean(mae_scores):.2f} Â± {np.std(mae_scores):.2f}")
print(f"Average RMSE: {np.mean(rmse_scores):.2f} Â± {np.std(rmse_scores):.2f}")
print(f"Average RÂ²: {np.mean(r2_scores):.4f} Â± {np.std(r2_scores):.4f}")
```
